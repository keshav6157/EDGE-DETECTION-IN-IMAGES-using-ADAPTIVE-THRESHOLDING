```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
org_img = cv2.imread("karim-manjra-4dPJQ6VsXjA-unsplash.jpg", 1)
plt.figure(figsize = (7,6))
plt.title("original image")
plt.imshow(org_img, 'gray')
```

```python
img = cv2.medianBlur(org_img, 5)
```

```python
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 2)
```

```python
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 9, 2)
```

```python
titles = ["median filter", "simple/global thresholding", "adaptive mean", "adaptive gaussian"]
```

```python
imgs = [img, th1, th2, th3]
plt.figure(figsize = (12, 10))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(imgs[i], "gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Result after otsu's thresholding
ret2, th4 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(th4, "gray")
```
